A recommender system with the given data using User Based Collaborative Filtering.  
This dataset is related to the video gaming industry and a survey was conducted to build a   
recommendation engine so that the store can improve the sales of its gaming DVDs.   
Following is a Recommendation Engine that suggests top selling DVDs based on which DVD the customer buys.

```python
gamemat = games.pivot_table(index='userId',columns='game',values='rating')
gamemat1= gamemat.fillna(0)
gamemat1.head()
```

```python
gamemat1.shape
```

```python
#create similarity matrix for users
from sklearn.metrics.pairwise import cosine_similarity
user_similarity_matrix = cosine_similarity(gamemat1)
```

```python
user_similarity_matrix
```

This is the matrix based on which we will recommmend games to users

```python
user_similarity_matrix.shape
```

```python
gamemat1.tail()
```

```python
#coonvert to datframe for ease of usage
user_similarity_df= pd.DataFrame(user_similarity_matrix,index=gamemat1.index,columns=gamemat1.index)
```

```python
user_similarity_df
```

```python
#get similar users where cosine_similarity > 0 
def get_similar_users(userId):
    idx = []
    if userId in user_similarity_df.index:
        similar_users = user_similarity_df[user_similarity_df[userId]>0][userId]
        similar_users = similar_users.index.drop(userId).sort_values(ascending=False)
    else:
        similar_users=np.array([])

    similar_users
    if similar_users.shape[0] :
        return similar_users.values
    else:
        return []
    
```

```python
#get games where similar users have rated more than 3
def get_recommended_games(users):
    return games[(games['userId'].isin(users)) & (games['rating']>3)]['game'].unique()    
```

```python
#get recommendations for user 2
get_recommended_games(get_similar_users(2))
```

```python
#get recommendations for user 6
get_recommended_games(get_similar_users(6))
```
