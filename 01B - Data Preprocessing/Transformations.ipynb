### Normalization

```python
### Normalization function - Custom Function
# Range converts to: 0 to 1
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
cal_norm = norm_func(cal)
cal_norm
```

```python
cal_norm.describe()
```

### Standardization

```python
### Standardization (z = (x - u) / s)
from sklearn.preprocessing import StandardScaler
# Initialise the Scaler
scaler = StandardScaler()
# To scale data
cal_std = scaler.fit_transform(cal)
# Convert the array back to a dataframe
cal_std_df = pd.DataFrame(cal_std)
cal_std_df.columns = cal.columns
cal_std_df.describe()
```
